# Steps:
- Read an image file.
- The image file will consist of RGB layers. So, convert all RGB layers to each vector. The image will change from (m,n,p) to (m*n,p).
- Pick some number of random centeroids (Using Elbow Method) and then radomly initialize them.
- Now, we have got the centroids. The next task is to optimize the centroids.
- So, for each data point, compute the euclidian distance from all the centroids and assign the cluster based on the minimal distance to all the centroids.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=mpimg.imread('index.jpeg')
```

```python
imgplot = plt.imshow(img)
```

```python
img=img/255
```

```python
img.shape
```

```python
img_new=np.reshape(img, (img.shape[0] * img.shape[1], img.shape[2]))
```

```python
m,n=img_new.shape
```

```python
def distance(x1,y1,z1,x2,y2,z2): 
    dist = np.sqrt(np.square(x1 - x2) + np.square(y1 - y2)+np.square(z1 - z2))
    return dist 
```

```python
centroids=np.zeros((16,n))# 16 is the number of colors we are choosing out of the image to be compressed
```

```python
# random initialization of Centroids.  
for i in range(16): 
    centroids[i] = img_new[int(np.random.random(1)*1000)]
```

```python
distance_array=np.zeros((m,1))
```

```python
centroids_dictionary={}
```

```python
def nearest_centroid():
    for i in range(img_new.shape[0]):
        dlist=[]
        for j in range(16):
            dlist.append(distance(img_new[i][0],img_new[i][1],img_new[i][2],centroids[j][0],centroids[j][1],centroids[j][2]))
        min_dist=np.argmin(dlist)
        centroids_dictionary[i]=min_dist   
    return centroids_dictionary
```

```python
comp = list(nearest_centroid().values())
```

```python
np.array(comp).shape
```

```python
len(centroids_dictionary)
```

```python
img_compressed=np.zeros(img_new.shape)
```

```python
for key, value in centroids_dictionary.items():
    img_compressed[key]=centroids[value]
```
